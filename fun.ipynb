```python
def float_to_scientific_notation(x, base=2):
  # Determine the sign
  if x < 0:
    s = -1
    x = -x
  else:
    s = 1

  # Determine the mantissa and exponent
  m = x
  e = 0
  while abs(m) < 1:
    m *= base
    e -= 1
  while abs(m) >= base:
    m /= base
    e += 1

  # Return the scientific notation representation as a string
  return f"{s}^s*(1+{m:.10f})*{base}^{e}"
```

```python
float_to_scientific_notation(15.25)
```
