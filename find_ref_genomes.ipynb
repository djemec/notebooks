```python
import requests
from bs4 import BeautifulSoup

core = 'https://ftp.ncbi.nlm.nih.gov/genomes/refseq/vertebrate_mammalian/'
la = 'latest_assembly_versions/'
end = 'genomic.fna.gz'
exclude = ['cds','rna']
```

```python
def ret_list(path, is_link = True):
    r = requests.get(path)
    soup = BeautifulSoup(r.text, "html.parser")
    links = []
    #print(soup)
    for a in soup.find_all("a")[1:]:
        href = a.get("href")
        if href is not None:
            if href.endswith('/') and is_link:
                links.append(href) 
            if href.endswith(end) and not is_link and all(x not in href for x in exclude):
                links.append(href)
    return links
```

```python
parent = ret_list(core)
parent
```

```python
end_file_path = [] 
for org in parent:
    print(org)
    # creates ../*/latest_assembly_versions/
    path = core + org + la
    #print(path)
    inter = ret_list(path)
    if inter:
        # creates .../latest_assembly_versions/*
        path_2 = path + inter[0]
        #print('p2: '+ path_2)
        inter_2 = ret_list(path_2, False)
        if inter_2:
            final_path = path_2 + inter_2[0]
            #print('final: ' + final_path)
            end_file_path.append(final_path)


end_file_path
```
