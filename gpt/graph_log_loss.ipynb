```python
from pathlib import Path
import matplotlib.pyplot as plt
```

```python
log_file = Path('/Users/djemec/Downloads/log.txt')
log2_file = Path('/Users/djemec/Downloads/log_restart.txt')
log3_file = Path('/Users/djemec/Downloads/log_restart2.txt')
```

```python
train = {'iter': [], 'loss': []}
test  = {'iter': [], 'loss': []}
```

```python
with open(log_file) as f:
    for line in f:
        it, kind, loss = line.split()
        target = test if kind == 'test' else train
        target['iter'].append(int(it))
        target['loss'].append(float(loss))
## add from restart
pad = max(train['iter']) +1
```

```python
with open(log2_file) as f:
    for line in f:
        it, kind, loss = line.split()
        target = test if kind == 'test' else train
        target['iter'].append(int(it) + pad)
        target['loss'].append(float(loss))

# repull
pad = max(train['iter']) +1
```

```python
with open(log3_file) as f:
    for line in f:
        it, kind, loss = line.split()
        target = test if kind == 'test' else train
        target['iter'].append(int(it) + pad)
        target['loss'].append(float(loss))

# repull
```

```python
plt.figure(figsize=(13,6))
plt.plot(train['iter'][::10], train['loss'][::10], label='train', lw=0.1)
plt.plot(test ['iter'], test ['loss'], label='test',  lw=2)

plt.xlabel('iteration')
plt.ylabel('loss')
#plt.yscale('log') 
plt.legend()
plt.ylim(2.8, 4)      

#plt.tight_layout()
plt.show()
```
